# Use pandas for feature encoding
---

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

### pandas get dummies function

```python
pd.get_dummies(df['day'])
```

```python
pd.get_dummies(df['day'], drop_first = True)
```

```python
get_dummies = pd.get_dummies(df, columns = ['day'])
get_dummies.head()
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns = ['day'], drop_first = True)
df.head()
```
